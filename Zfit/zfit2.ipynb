```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import zfit

#Creating the space
obs_x = zfit.Space('x', limits=(-10, 10))
obs_y = zfit.Space('y', limits=(-10, 10))

#Parameters
mu = zfit.Parameter("mu", 1.0, -4, 6)
sigma = zfit.Parameter("sigma", 1., 0.1, 10)
lambd = zfit.Parameter("lambda", 1., -5., 2)
```

```python
print(np.random.normal())
```

```python
#Now, we create these PDF'S with our previous parameters
gauss = zfit.pdf.Gauss(mu=mu, sigma=sigma, obs=obs_x)
exp = zfit.pdf.Exponential(lambda_=lambd, obs=obs_y)

model = zfit.pdf.ProductPDF(pdfs=[gauss,exp])

# data
#gauss_data = gauss.sample(1000).numpy()
#exp_data = exponential.sample(1000).numpy()

data = model.create_sampler(n=1000, fixed_params=True)

#data_g = model.create_sampler(1000, limits=obs_x)
#data_e = model.create_sampler(1000, limits=obs_y)
#data_g.resample()
#data_e.resample()

#mu.set_value(0.5)
#sigma.set_value(1.2)
#lambd.set_value(-0.05)
```

```python
#Create NLL
nll = zfit.loss.UnbinnedNLL(model=model, data=data)
#nll1 = zfit.loss.UnbinnedNLL(model=gauss, data=data_g)
#nll2 = zfit.loss.UnbinnedNLL(model=exp, data=data_e)

#Create the minimizer
minimizer = zfit.minimize.Minuit()

#result_g = minimizer.minimize(nll1)
#result_e = minimizer.minimize(nll2)
```

```python
listmu=[]
listsig=[]
listlam=[]
mu.set_value(1.0)
sigma.set_value(1.0)
lambd.set_value(1.0)
#Generating 100 samples
for i in range(100):
    #Initialize parameters randomly
    data.resample()
    
    #mu.set_value(abs(np.random.normal()))
    #sigma.set_value(abs(np.random.normal()))
    #lambd.set_value(abs(np.random.normal()))
    
    print("mu = ", mu.numpy())
    print("sigma = ", sigma.numpy())
    print("lambda = ", lambd.numpy())
        
    listmu.append(mu.numpy())
    listsig.append(sigma.numpy())
    listlam.append(lambd.numpy())
    
    result = minimizer.minimize(nll)
    
```
